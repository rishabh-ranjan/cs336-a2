```python
from collections import defaultdict
import expt
import pandas as pd
```

```python
runs_dir = "/dfs/scratch1/ranjanr/runs/cs336/2024-05-04_ddp"
df = expt.scan(runs_dir)
df
```

```python
def make_pd_tab(local_world_size, no_ddp):
    tab = defaultdict(dict)
    for lm_size in [
        "small",
        "medium",
        "large",
        "xl",
    ]:
        if no_ddp:
            ts = df.query("lm_size == @lm_size and ddp == 0").index[-1]
            tab[lm_size]["no ddp"] = df["report_time"][ts]
            continue

        ts = df.query("lm_size == @lm_size and local_world_size == @local_world_size and naive == 1").index[-1]
        tab[lm_size]["naive ddp"] = df["report_time"][ts]
        
        ts = df.query("lm_size == @lm_size and local_world_size == @local_world_size and bucket_size_mb == inf").index[-1]
        tab[lm_size]["flat ddp"] = df["report_time"][ts]
        
        ts = df.query("lm_size == @lm_size and local_world_size == @local_world_size and bucket_size_mb == 0").index[-1]
        tab[lm_size]["overlap idv params"] = df["report_time"][ts]

        for bucket_size_mb in [
            5,
            10,
            50,
            100,
            500,
        ]:
            ts = df.query("lm_size == @lm_size and local_world_size == @local_world_size and bucket_size_mb == @bucket_size_mb").index[-1]
            tab[lm_size][f"buckets of {bucket_size_mb} MB"] = df["report_time"][ts]
    pd_tab = pd.DataFrame(tab)
    return pd_tab
```

```python
for local_world_size in [
    2,
    1,
]:
    print(f"{local_world_size=}")
    for no_ddp in [True, False]:
        pd_tab = make_pd_tab(local_world_size=local_world_size, no_ddp=no_ddp)
        display(pd_tab.style.format("{:.3f}").background_gradient(axis=0))
```
